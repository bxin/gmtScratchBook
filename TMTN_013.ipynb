```python
dfs = pd.read_excel('TMTN-013-data.xlsx', sheet_name='SMR Translation Test', header=1)
```

```python
x = np.array(dfs['x'][:21])
contrast = np.array(dfs['contrast'][:21])
m = np.array(dfs['diff'][:21])
n = len(x)
print(n)
```
